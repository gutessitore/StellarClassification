### Pré processamento/Tratamento dos dados

```python
df.head()
```

```python
#transformando a coluna 'class' em numérica usando onehotenconding

df_str = df['class'].to_frame() #selecionando apenas a coluna numérica

encoder = OneHotEncoder()

df_str = encoder.fit_transform(df_str).toarray()
df_str
```

```python
#transformando o onehotencoding em um frame para inseri-lo no dataset

df_str = pd.DataFrame(df_str) #transformando em df

df_all = pd.concat([df, df_str], axis=1) #inserindo no dataset
df_all = df_all.drop(['class', 'rerun_ID', 'obj_ID'], axis=1) #rerun_ID possui 100.000 valores iguais
                                                              #obj_ID é apenas um nº identificação de objeto feita pela CAS
df_all
```

###### 1 0 0 - GALAXY

###### 0 1 0 - QUASAR

###### 0 0 1 - STAR

```python
#normalização dos dados antes do treinamento e teste

X = df_all.drop([0, 1, 2], axis=1) #features
y = df_all[[0, 1, 2]]              #target

scaler = StandardScaler()

X = scaler.fit_transform(X)
X
```

```python
#visualizando no dataset
pd.DataFrame(X)
```

```python
#No formato acima, os dados ficam em um formato de difícil análise, assim aplica-se o inverse_transform(), que faz com que os dados 
#já normalizados voltam a ficar no formato original a fim de uma melhor visualizção e/ou análise

X = scaler.inverse_transform(X)
pd.DataFrame(X)

#agora os dados estão mais parecidos com o formato original e também normalizados
```

### Treinamento e teste

```python
# X --> features 
# y --> target 
#já especficados acima

cl_tree = DecisionTreeClassifier()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
cl_tree = cl_tree.fit(X_train, y_train)
cl_tree
```

```python
previsao = cl_tree.predict(X_test)
previsao
```

```python
print(classification_report(y_test, previsao))
```
